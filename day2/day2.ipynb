```python
import numpy as np
```

```python
# A - rock, B - paper, C - scissors
# X - rock, Y - paper, Z - scissors
def determine_outcome(play):
    match play :
        case 'ay' | 'bz' | 'cx':
            return 6
        case 'ax' | 'by' | 'cz':
            return 3
        case _ :
            return 0

action_score = {
    'x': 1,
    'y': 2,
    'z': 3,
}
```

```python
res_list = []
with open('day2_input.txt') as f:
    for line in f:
        f_line = line.replace(" ","").replace("\n","").lower()
        print(f_line)
        print(action_score[f_line[1]])
        res_list.append(determine_outcome(f_line) + action_score[f_line[1]])
sum(res_list)
```

```python
# part 2
# X - Lose, Y - Tie, Z - Win
play_action_dict = {
    'x': 0,
    'y': 3,
    'z': 6
}

def determine_action_played_score(play):
    match play:
        case "ay" | "cz" | "bx":
            return 1
        case "az" | "by" | "cx":
            return 2
        case "ax" | "bz" | "cy":
            return 3
        case _:
            return 0;
```

```python
res_list2 = []
with open('day2_input.txt') as f:
    for line in f:
        f_line = line.replace(" ","").replace("\n","").lower()
        print(f_line)
        print(play_action_dict[f_line[1]])
        res_list2.append(determine_action_played_score(f_line) + play_action_dict[f_line[1]])
sum(res_list2)
```
